> This notebook downloads the latest time series from FRED and constructs the dataset that can be passed to the pymc model

-------------------------

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt 
import scipy.io as sio
import pandas as pd
import pandas_datareader.data as web
```

```python
# Download the latest monthly data from FRED
cpi_monthly = web.DataReader('CPIAUCSL', 'fred', start = pd.datetime(1948, 1, 1))
ur_monthly = web.DataReader('UNRATE', 'fred',  start = pd.datetime(1948, 4, 1))
tb3_monthly = web.DataReader('TB3MS', 'fred',  start = pd.datetime(1948, 4, 1))

# Construct the quarterly sample
cpi = np.log(cpi_monthly.groupby([cpi_monthly.index.year, 
                                  cpi_monthly.index.quarter]).last()).diff()[1:]

tb3 = tb3_monthly.groupby([tb3_monthly.index.year, 
                           tb3_monthly.index.quarter]).first()/400

ur = ur_monthly.groupby([ur_monthly.index.year, 
                         ur_monthly.index.quarter]).mean()/100
```

```python
date_index = pd.date_range('1948-04-01', '2016-07-01', freq = 'Q')
data_updated = pd.DataFrame(data = {'dp' : cpi['CPIAUCSL'], 
                                    'y3' : tb3['TB3MS'], 
                                    'ur' : ur['UNRATE'] })
data_updated.index = date_index
```

```python
data_old = sio.loadmat('../data/NEWQDATA.MAT')

var_names = [key_name for key_name in data_old.keys() if key_name[0] != '_']
d = {}
for i in var_names:
    d[i] = data_old[i].squeeze()
data_old = pd.DataFrame(data = d, index = pd.date_range('1948-04-01', '2000-12-31', freq = 'Q'))
```

### Compare the new to the old series

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 10))
data_updated['dp'].plot(ax = ax[0, 0], label = 'updated')
data_old['dp'].plot(ax = ax[0, 0], label = 'old')
ax[0, 0].set_title('cpi')
ax[0, 0].legend(loc = 'best')

data_updated['ur'].plot(ax = ax[0, 1], label = 'updated')
data_old['ur'].plot(ax = ax[0, 1], label = 'old')
ax[0, 1].set_title('unemp rate')

data_updated['y3'].plot(ax = ax[1, 0], label = 'updated')
data_old['y3'].plot(ax = ax[1, 0], label = 'old')
ax[1, 0].set_title('t-bill')
```

### Save the updated dataset as a csv file

```python
data_updated.to_csv("../data/updated_sample.csv", sep = ',')
```
